```python
# importing essential libraries
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("Health_heart_experimental.csv")
df.head(10)
```

#### Age Vs Height Visualization of the given dataset.

```python
sns.lineplot(data= df , x = 'age' , y = 'heightCm', hue= 'sex')
plt.title("Age Vs Height (Cm) Chart")
```

#### Comparison of weights (Kg) of Males and Females on the basis of their age. 

```python
sns.scatterplot(data= df.head(200) , x = 'age', y = 'weightKg', hue='sex')
```

```python
sns.barplot(data= df.head(100) , x='SysBP', y = 'DiaBP',hue= 'sex')
```
