## Question -1
Descriptive Statistics 

Use a dataset : Iris Dataset. Perform the following tasks:    

 a. Compute the mean, median, and standard deviation for all numerical features. 


 b. Create histograms and boxplots for feature distributions.


 c. Calculate the correlation matrix for the features.

```python
df = pd.read_csv("iris.csv")
print(df.head())

print("\n")

# computing means ...
mean_sepalLength = np.mean(df['sepal.length'])
mean_sepalWidth = np.mean(df['sepal.width'])
mean_petalLength = np.mean(df['petal.length'])
mean_petalWidth = np.mean(df['petal.width'])

# Printing means ...
print("Mean Sepal_Length = ",round(mean_sepalLength,4))
print("Mean Sepal_Width = ",round(mean_sepalWidth,4))
print("Mean Petal_Length = ",round(mean_petalLength,4))
print("Mean Petal_Width = ",round(mean_petalWidth,4))

# computing medians ...
median_sepalLength = np.median(df['sepal.length'])
median_sepalWidth = np.median(df['sepal.width'])
median_petalLength = np.median(df['petal.length'])
median_petalWidth = np.median(df['petal.width'])

print("\n")

# Printing medians ...
print("Median Sepal_Length = ",round(median_sepalLength,4))
print("Median Sepal_Width = ",round(median_sepalWidth,4))
print("Median Petal_Length = ",round(median_petalLength,4))
print("Median Petal_Width = ",round(median_petalWidth,4))

# computing standard deviation ...
std_sepalLength = np.std(df['sepal.length'])
std_sepalWidth = np.std(df['sepal.width'])
std_petalLength = np.std(df['petal.length'])
std_petalWidth = np.std(df['petal.width'])

print("\n")

# Printing standard deviation ...
print("Standard Deviation Sepal_Length = ",round(std_sepalLength,4))
print("Standard Deviation Sepal_Width = ",round(std_sepalWidth,4))
print("Standard Deviation Petal_Length = ",round(std_petalLength,4))
print("Standard Deviation Petal_Width = ",round(std_petalWidth,4))
```

## Question -3
-> Bayes' Theorem


A model classifies spam emails with the following probabilities:


P(Spam)=0.2


P(Not Spam)=0.8


P(Keyword | Spam)=0.7


P(Keyword | Not Spam)=0.1


Compute the probability that an email is spam given that it contains the keyword.

```python
P_spam = 0.2
P_notSpam = 0.8
P_keyword_givenThat_spam = 0.7
P_keyword_givenThat_notSpam = 0.1

# calculating P(Keyword and Spam) ..
P_keyword_and_spam = P_keyword_givenThat_spam * P_spam

# calculating P(Keyword) using bayes theorem ..
P_keyword = (P_keyword_givenThat_spam * P_spam) + (P_keyword_givenThat_notSpam * P_notSpam)

# calculating P(Spam | Keyword) ..
P_spam_givenThat_keyword = P_keyword_and_spam / P_keyword
print("P(Spam|Keyword) = ", round(P_spam_givenThat_keyword,3))
```
